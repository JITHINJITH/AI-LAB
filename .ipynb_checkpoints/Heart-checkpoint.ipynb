```python
import numpy as np
import pandas as pd
```

```python
heart=pd.read_csv('heart.csv')
heart.head(10)
```

```python
x=heart[['trtbps','chol','thalachh','oldpeak']]
x.head(10)
```

```python
y=heart['output']
y.head(10)
```

```python
from sklearn.svm import SVC
ML=SVC()
ML=ML.fit(x,y)
```

```python
op=ML.predict([[ 130, 240, 120, 3.7]])
print(op)
```

```python
from sklearn.naive_bayes import GaussianNB
ML=GaussianNB()
ML=ML.fit(x,y)
```

```python
op=ML.predict([[ 130, 240, 120, 3.7]])
print(op)
```
