```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
mall = pd.read_csv('/content/Mall_Customers.csv')
x = mall.iloc[:,[3,4]].values
```

```python
from scipy.cluster import hierarchy as hc
z = hc.linkage(x,method = 'ward')
dn = hc.dendrogram(z)
plt.xlabel('customer')
plt.ylabel('euclidean distance')
plt.title('hierarchical clustering')
```

```python
from sklearn.cluster import AgglomerativeClustering
ag = AgglomerativeClustering(n_clusters=5,affinity='euclidean',linkage='ward')
y_ag = ag.fit_predict(x)
y_ag
```

```python
plt.scatter(x[y_ag ==0,0],x[y_ag==0,1],s = 100,c ='blue',label = 'cluster1')
plt.scatter(x[y_ag ==1,0],x[y_ag==1,1],s = 100,c ='green',label = 'cluster2')
plt.scatter(x[y_ag ==2,0],x[y_ag==2,1],s = 100,c ='orange',label = 'cluster3')
plt.scatter(x[y_ag ==3,0],x[y_ag==3,1],s = 100,c ='red',label = 'cluster4')
plt.scatter(x[y_ag ==4,0],x[y_ag==4,1],s = 100,c ='cyan',label = 'cluster5')
plt.legend()
plt.xlabel('income')
plt.ylabel('spending rate')
plt.title('hierarchical clustering')
plt.show()
```
